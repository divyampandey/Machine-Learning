```python
import pandas as pd

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"

col_list=['sepal_length','sepal_width','petal_length','petal_width','target']

df=pd.read_csv(url,names=col_list)
```

```python
df.info()

```

```python
from sklearn.preprocessing import StandardScaler
```

```python
features=df.columns.tolist()[:-1]
```

```python
features
```

```python
x=df.loc[:,features].values
y=df.loc[:,['target']].values
```

```python
x
```

```python
y
```

```python
x=StandardScaler().fit_transform(x)
```

```python
x
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2)
```

```python
principalcomponents=pca.fit_transform(x)
```

```python
principledf=pd.DataFrame(data=principalcomponents, columns=['principle comp 1', 'principle comp 2'])
```

```python
principledf
```

```python
finaldf=pd.concat([principledf,df[['target']]],axis=1)
```

```python
finaldf
```

```python
colors={'Iris-setosa':'r','Iris-versicolor':'g','Iris-virginica':'b'}
mydf=pd.DataFrame([colors])
```

```python
def assign_color(var):
        return colors[var]
```

```python
finaldf['color']=df['target'].apply(assign_color)
```

```python
mycol=finaldf['color'].values
mycol
```

```python
xs=finaldf.loc[:,['principle comp 1']]
ys=finaldf.loc[:,['principle comp 2']]

import matplotlib.pyplot as plt
plt.scatter(xs,ys,alpha=0.5,c=mycol)
plt.legend()

plt.show()
```
